```python
import math

def multiply(x,y):
    if len(x)==1 and len(y)==1:
        return str(int(x[0])*int(y[0]))
    else:
        # (ab)(cd) = ac*10^(2*half) + (ad+bc)*10^(half) + bd
        n = max(len(x),len(y))
        half = n//2
        a = x[:-half]
        b = x[-half:]
        c = y[:-half]
        d = y[-half:]
        ac = multiply(a,c) + '0'*(2*half)
        ad = multiply(a,d)
        bc = multiply(b,c)
        ad_plus_bc = str(int(ad) + int(bc))+'0'*(half)
        bd = multiply(b,d)
        return str(int(ac) + int(ad_plus_bc) + int(bd))
```

```python
import time
import random

n = 2
high = 90
low = 0
up_to = 2048

while n <= up_to:

    x = [random.randint(low, high) for _ in range(n)]
    y = [random.randint(low, high) for _ in range(n)]
    t_start = time.time_ns()
    xy = multiply(x,y)
    t_stop = time.time_ns()
    t = t_stop - t_start
    print(f"{n:6d} \t{t:14,d}")
    n *= 2
```
